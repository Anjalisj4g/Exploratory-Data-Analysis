```python
#1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.
```

```python
#3. Identify the predominant age group among employees. 
```

```python
df['Age'].value_counts()
```

```python
age_groups = [18, 25, 30, 35, 42] 

labels = [f'{start}-{end-1}' for start, end in zip(age_groups[:-1], age_groups[1:])]
# Last group goes up to infinity
labels.append(f'{age_groups[-1]}+')

# Cut the ages into age groups
df['Age Group'] = pd.cut(df['Age'], bins=age_groups, labels=labels[:-1], right=False)

# Count occurrences of each age group
age_group_counts = df['Age Group'].value_counts()

# Find the predominant age group
predominant_age_group = age_group_counts.idxmax()
print("Predominant Age Group:", predominant_age_group)
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df.dropna(subset=['Age'], inplace=True)

plt.hist(df['Age'], bins=[18, 25, 30, 35, 42],color='yellow',edgecolor='b')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.title('Histogram of Employee Ages')
plt.show()
```

```python
Employees of age group 25-30 are the predominant group followed by age group 18-25. Employees of age group 35-40 are least dominant in the data.
```

```python
#4. Discover which team and position have the highest salary expenditure.
```

```python
grouped_by_team = df.groupby('Team')['Salary'].sum().reset_index()
grouped_by_position = df.groupby('Position')['Salary'].sum().reset_index()

max_salary_team = grouped_by_team.loc[grouped_by_team['Salary'].idxmax()]
max_salary_position = grouped_by_position.loc[grouped_by_position['Salary'].idxmax()]

print("Team with Highest Salary Expenditure is",max_salary_team['Team'])
print("Total Salary Expenditure:", max_salary_team['Salary'])

print("\nPosition with Highest Salary Expenditure is",max_salary_position['Position'])
print("Total Salary Expenditure:", max_salary_position['Salary'])
```

```python
grouped_by_team = df.groupby('Team')['Salary'].sum().reset_index()
grouped_by_position = df.groupby('Position')['Salary'].sum().reset_index()

max_salary_team = grouped_by_team.loc[grouped_by_team['Salary'].idxmax()]
max_salary_position = grouped_by_position.loc[grouped_by_position['Salary'].idxmax()]

plt.figure(figsize=(10, 6))
plt.bar(grouped_by_team['Team'], grouped_by_team['Salary'],color='lightgreen',edgecolor='blue')
plt.xlabel('Team')
plt.ylabel('Total Salary Expenditure')
plt.title('Total Salary Expenditure by Team')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(grouped_by_position['Position'], grouped_by_position['Salary'],color='pink',edgecolor='blue')
plt.xlabel('Position')
plt.ylabel('Total Salary Expenditure')
plt.title('Total Salary Expenditure by Position')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
From the graph, Team with Highest Salary Expenditure is Memphis Grizzlies and team with Least Salary Expenditure is Orlando Magic.
Also, Position with Highest Salary Expenditure is PF and position with least Salary Expenditure is C.
```

```python
#5. Investigate if there's any correlation between age and salary, and represent it visually.
```

```python
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df.dropna(subset=['Salary'], inplace=True)
df1=df.drop(['Name','Team','Position','College','Age Group'],axis='columns')


corr=df1.corr()
sns.heatmap(corr,annot=True,cmap="coolwarm") 
plt.title('Correlation Matrix of Employee Details of ABC Company')
plt.show()
```

```python
From the correlation matrix, the correlation coefficient of age and salary is exactly 1. 
This means that there is a strong positive linear relationship between these two variables.
This indicates that as age increases, salary also increases proportionally and vice-versa. 
Through this relationship, we can predict the salary of an employee by knowing his/her age. 
```
